# The Birds and the Planes

With the complexity of airports and the long lines and infamy of airplane food, it is often easy to overlook some large concerns dealing with airports and airlines. Birdstrikes are an example of a serious issue with varying consequences, whether it's the possibility of birds clogging an engine and causing a plane to be unable to operate, or an issue with the delicate balance of the environment. 

Wikipedia has a list of the world's busiest airports, which would most likely be a destination for a traveler at some point. <br/>

The FAA has a database of every documented instance of wildlife interference available for download. <br/>
http://wildlife.faa.gov/

With the potential for birds to be so dangerous around airplanes, of those ten cities, my focus is to figure out which ones are the best and worst when it comes to the rate of accidents involving wild animals.

```python
import csv
from collections import Counter
from collections import defaultdict
import re

def readCSV(filename):
    with open(filename) as f:
        arr = [line for line in csv.reader(f)]
    return arr

atl = readCSV("atl.csv")
lax = readCSV("klax.csv")
ord_airport = readCSV("ord.csv") #ord happens to be a reserved word
dfw = readCSV("dfw.csv")
den = readCSV("den.csv")
jfk = readCSV("jfk.csv")
sfo = readCSV("sfo.csv")
clt = readCSV("clt.csv")
las = readCSV("las.csv")
phx = readCSV("phx.csv")
```

If you check the csv's for each of the airports (they're labeled by airport code), you will probably notice that the number of birdstrikes may vary, and not all these incidents are in fact birds (there are woodchucks and deer also on file). Additionally, you will probably notice that some of the birdstrikes aren't very clear on quantity of birds being hit. For example, a lot of them will have "2 to 10 listed" for the number of birds hit. After much deliberation, I concluded that in counting the number of incidents, the number of birds involved wouldn't actually affect the results, as whether it's one bird or two flying side by side, a hit is still a hit. 

On to the next step: Counting the frequency of planes that fly out of each airport. Since many more planes fly out of Atlanta than out of Phoenix, it would be inaccurate to just compare birdstrike frequencies straight. I'd be better off mapping birdstrikes per takeoff and landing (a plane mid-flight would be too high to hit a bird).

```python
def parse_year(date):
    return re.sub("[$[0-9]+/","",date)[:4]
```

```python
def count_incidents(airport):
    airport = [line[0] for line in airport]
    airport.remove('INCIDENT_DATE') #the title needs to be removed to continue
    airport = [parse_year(line) for line in airport]
    
    return Counter(airport)
```

```python
atl_counts = count_incidents(atl)
lax_counts = count_incidents(lax)
ord_counts = count_incidents(ord_airport)
dfw_counts = count_incidents(dfw)
den_counts = count_incidents(den)
jfk_counts = count_incidents(jfk)
sfo_counts = count_incidents(sfo)
clt_counts = count_incidents(clt)
las_counts = count_incidents(las)
phx_counts = count_incidents(phx)
```

Now that we have the number of incidents per year, it's time to normalize it against the number of planes that took off and landed. Luckily, the government releases those statistics too!

```python
operations = defaultdict(dict)

with open("flying_freq.csv") as f:
    f = [line for line in csv.reader(f)]
```

```python
f.remove(f[0])
```

```python
for line in f:
    operations[line[0]][line[1]] = float(re.sub(",","",line[2]))
```

```python
atl_counts = {key: float(atl_counts[key])/operations["ATL"][key] for key in operations["ATL"].viewkeys() & atl_counts.viewkeys()}
```

```python
def normalize(airport_data, airport):
    return {key: float(airport_data[key])/operations[airport][key] for key in operations[airport].viewkeys() & airport_data.viewkeys()}
```

```python
n_atl = normalize(atl_counts,"ATL")
n_lax = normalize(lax_counts,"LAX")
n_ord = normalize(ord_counts,"ORD")
n_dfw = normalize(dfw_counts,"DFW")
n_den = normalize(den_counts,"DEN")
n_jfk = normalize(jfk_counts,"JFK")
n_sfo = normalize(sfo_counts,"SFO")
n_clt = normalize(clt_counts,"CLT")
n_las = normalize(las_counts,"LAS")
n_phx = normalize(phx_counts,"PHX")
```

```python
total_atl = sum(n_atl.values())
total_lax = sum(n_lax.values())
total_ord = sum(n_ord.values())
total_dfw = sum(n_dfw.values())
total_den = sum(n_den.values())
total_jfk = sum(n_jfk.values())
total_sfo = sum(n_sfo.values())
total_clt = sum(n_clt.values())
total_las = sum(n_las.values())
total_phx = sum(n_phx.values())
```

```python
import json

dataset = json.dumps({"atl":total_atl,"lax":total_lax,"ord":total_ord,"dfw":total_dfw,"den":total_den,"jfk":total_jfk,"sfo":total_sfo,"clt":total_clt,"las":total_las,"phx":total_phx})
```
