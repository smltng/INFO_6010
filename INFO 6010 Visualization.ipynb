```python
import csv
from collections import Counter
from collections import defaultdict
import re

def readCSV(filename):
    with open(filename) as f:
        arr = [line for line in csv.reader(f)]
    return arr

airports = ["atl","klax","ord","dfw","den","jfk","sfo","clt","las","phx"]

raw_data = defaultdict(list)

for airport in airports:
    raw_data[airport] = readCSV(airport+".csv")
```

If you check the csv's for each of the airports (they're labeled by airport code), you will probably notice that not all these incidents are in fact birds (there are woodchucks and deer also on file, for example). Additionally, you will probably notice that some of the birdstrikes aren't very clear on quantity of birds being hit. For example, a lot of them will have "2 to 10 listed" for the number of birds hit. After much deliberation, I concluded that in counting the number of incidents, the number of birds involved isn't of as much importance. Since a plane can be quite large, it is tough for birds to avoid planes, whether they are in a group or individual, which may lead to a strike. 

On to the next step: Counting the frequency of planes that fly out of each airport. Since many more planes fly out of Atlanta than out of Phoenix, it would be inaccurate to just compare birdstrike frequencies straight. I'd be better off mapping birdstrikes per takeoff and landing (a plane mid-flight would be too high to hit a bird).

```python
def parse_year(date):
    return re.sub("[$[0-9]+/","",date)[:4]
```

```python
def species_found(airport):
    species = [line[9] for line in airport]
    species = Counter(species)
    
    return species.keys()
```

In parsing the species found in the CSV files, it quickly becomes obvious that not all of these incidents occur with birds. For the sake of clarity, I parsed through each of these and identified the ones that weren't birds.

```python
species_found(raw_data["ord"])
```

These are the elements that aren't birds:

```python
non_birds = ['Virginia opossum', 'Red bat', 'Common gray fox', 'Little brown bat', 'SPECIES', 'Coyote', 'Microbats', 'Free-tailed bats', 'Domestic cat', 'Domestic dog', 'Striped skunk', 'Raccoon', 'Hoary bat', 'Brazilian free-tailed bat', 'Red fox',  'Muskrat', 'Woodchuck','White-tailed deer','Foxes','Indiana bat','Turtles','Rabbits','Black-tailed jackrabbit','Skunks', 'Hares','Badger','Black-tailed prairie dog','Big brown bat', 'Unknown terrestrial mammal','Silver-haired bat','Free-tailed bats','Vesper bats','Common snapping turtle', 'Kit fox', ]
```

```python
def count_incidents(airport):
    airport = [line[0] for line in airport if line[9] not in non_birds]
    airport = [parse_year(line) for line in airport]
    
    return Counter(airport)
```

```python
counts = defaultdict(dict)

for airport in airports:
    counts[airport] = count_incidents(raw_data[airport])
```

Now that we have the number of incidents per year for the past 15 years, it's time to normalize it against the number of planes that took off and landed. Luckily, the government releases those statistics too!

```python
operations = defaultdict(dict)

with open("flying_freq.csv") as f:
    f = [line for line in csv.reader(f)]
```

```python
f.remove(f[0])
```

```python
for line in f:
    operations[line[0]][line[1]] = float(re.sub(",","",line[2]))
```

```python
atl_counts = {key: float(counts["atl"][key])/operations["ATL"][key] for key in operations["ATL"].viewkeys() & counts["atl"].viewkeys()}

atl_counts
```

```python
def normalize(airport_data, airport):
    return {key: float(airport_data[key])/operations[airport][key] for key in operations[airport].viewkeys() & airport_data.viewkeys()}
```

```python
normalized = defaultdict(dict)
for airport in airports:
    normalized[airport] = normalize(counts[airport],airport.upper())
```

```python
total = defaultdict(dict)

for airport in airports:
    total[airport] = sum(normalized[airport].values())
```

## Part 2: Migration Patterns

Let's take a deeper look into birdstrike frequency in relation to migration patterns. With the unexpected JFK results, my next point of interest is how much migration patterns actually matter. One picture from Cornell's Lab of Ornithology shows the migratory pattern of birds.

```python
Image(url="http://www.birds.cornell.edu/AllAboutBirds/studyingbirdsi/bobolink_with_map1.jpg")
```

On the map, we clearly see that birds will only stay in JFK and ORD for the summer, so those two places are especially of interest. I would expect to see a higher frequency in birdstrikes for JFK and ORD for a couple reasons. One, if there are fewer birds in the winter, there would naturally be fewer chances of birdstrikes occurring, making it less likely for a higher number of birdstrikes during that season. Secondly, due to the weather, flight cancellations occur more frequently during the winter, meaning fewer flights and less birdstrike chances.

```python
def parse_date(date):
    return re.sub("/[0-9]+/[0-9]+ 0:00","",date)

#should return 8 (August)
parse_date(raw_data["atl"][1][0])
```

```python
aggregate = defaultdict(dict)
for airport in airports:
    aggregate[airport] = defaultdict(dict)
    
for airport in airports:
    for entry in range(1,len(raw_data[airport])):
        #empty dictionaries evaluate to false
        if not aggregate[airport][parse_date(raw_data[airport][entry][0])]:
            aggregate[airport][parse_date(raw_data[airport][entry][0])] = 1
        else:
            aggregate[airport][parse_date(raw_data[airport][entry][0])] += 1
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def to_tuple(airport):
    results = []
    for i in range(1,13):
        results.append(aggregate[airport][str(i)])
    return tuple(results)

def plot_strikes(airport):
    N = 12
    ind = np.arange(N)
    width = .35

    fig,ax = plt.subplots()
    rects = ax.bar(ind, to_tuple(airport), width, color='blue')

    ax.set_ylabel('Total Birdstrike Occurrences')
    ax.set_title('Aggregate Birdstrikes At '+airport.upper()+' by Month (1990-2014)')
    ax.set_xticks(ind+width)
    ax.set_xticklabels( ('Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec') )

    plt.show()
    
```

```python
for airport in airports:
    plot_strikes(airport)
```

As expected of ATL (Atlanta) and CLT (Charlotte), as the birds start to migrate south for the winter, birdstrikes become more frequent, reaching a peak in September and October, which is when the birds are expected to pass through. The frequency drops during the winter months before slowly climbing back up as the birds migrate back, although the numbers are nowhere near as pronounced as they were for the south-bound migration. ORD and JFK's results overall seem to match the hypothesis as well, with the winter months having fewer strikes in comparison. The large number of birdstrikes at JFK seems a little out of place, however, as I would have expected a good portion of the birds to have already left for warmer weather. Another really odd case is DEN, which has pronouncedly fewer birdstrikes in the winter months before the number skyrockets during a couple summer months...
