We begin, as usual, by importing a bunch of necessary tools for the job. In this case, it's the API for NYT, NLTK, and Numpy. I also import the sleep function as a matter of courtesy to NYT as I'm going to be making a lot of calls...

```python
from nytimesarticle import articleAPI
from time import sleep
import nltk
import numpy
```

```python
#You'll have to use your own API key here...
api = articleAPI('xxxxxxxxxxxxxxx')
```

```python
articles = []

#0 to 20 throws an error so it only goes from 0 to 19
for i in numpy.arange(0,19):
    articles.append(api.search( q = 'Obama', fq = {'headline':'Obama', 'source':['Reuters','AP', 'The New York Times']}, begin_date = 20090120, page = i, sort = 'oldest', facet_field = ['source','day_of_week'], facet_filter = True ))
    sleep(1) #pause for one second to give the server breathing room
```

After a little struggle, I finally figured out how the dataset is structured. Each article returned by the API contains the headlines of the article, byline (which contains the author), and snippets of the actual article. To test, I grab all the headlines about articles mentioning Obama.

```python
for article in articles:
    for docs in article['response']:
        for entry in article['response'][docs]:
            if type(entry) is dict:
                print entry['headline']['main']
```

You'd think that I was done with my data curation at this point, but the worst stage of the cleaning had yet to come. After some thought, I decided that I didn't want to rush into the dataset without getting a feel for what was available. Looking through the snippets of the original article seemed like a good place to begin. I ended up going with counting the frequencies of various words in the document and using those to try to see if I could draw anything useful. If not, I would at least have a hint of what other things to try before I fully committed to something more time consuming that could potentially yield little to no results...

On to my huge problem. So it turns out the data I grabbed is all in Unicode. "So what?", you ask? I kept getting a strange error saying "'ascii' codec can't encode character". Embarrassingly enough, it took me what seemed like ages to realize the differences between UTF-8 and Unicode encodings were making my code break. Turns out wrapping everything in unicode() seemed to fix the issue...

```python
entries = unicode("")
for article in articles:
    for docs in article['response']:
        for entry in article['response'][docs]:
            if type(entry) is dict:
                #compile the snippets into a list
                entries = unicode(entries) + unicode(" ") + unicode(entry['lead_paragraph'])
```

With the list of words finally set up, I can actually go count the words. Just kidding, there's one more important thing to do. I'm nearly certain the most common word I'm going to see is "a" or "the" if I run it on this dataset, so I choose to remove stopwords from the dataset. After some preliminary runs of the code, I realized that "President", "Barack", and "Obama" were there despite not really being all that interesting (the reasoning behind this being that I would of course find those words if I'm looking for Obama's name in a corpus).

```python
from nltk.corpus import stopwords

stop = stopwords.words('english')
stop+=["president","barack","obama"]
stopped_set = [word for word in entries.split() if word.lower() not in stop]
```

```python
from collections import Counter

Counter(stopped_set).most_common(50)
```

Of the most common 50 words aggregated across synposes of NYT articles since Obama's inauguration in 2009, we see a couple things are indeed evident. The word "new" appears quite a number of times, which would make sense if we think about Obama's initial campaign centered around "change". If the word "new' is appearing in the news, it could be an indication that change indeed has been happening. Other interesting words appear, such as "economic" and "stimulus", which hint at our need for an improved economy--something else the news would definitely hone in on.
